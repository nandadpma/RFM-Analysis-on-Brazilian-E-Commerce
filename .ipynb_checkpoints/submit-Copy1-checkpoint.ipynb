```python
customer_order_freq = new_order.groupby(['customer_id']).agg(order_freq=('order_id','nunique')).reset_index()
customer_total_spending = new_order.groupby(['customer_id']).agg(total_spending=('payment_value','sum')).reset_index()
customer_order_details = pd.merge(customer_order_freq, customer_total_spending, on='customer_id', how='left')

new_order['order_purchase_timestamp'] = pd.to_datetime(new_order.order_purchase_timestamp)
recent_order = new_order.groupby(['customer_id']).agg(recent_order=('order_purchase_timestamp','max')).reset_index()
recent_order['days_recency'] = (new_order.order_purchase_timestamp.max()-recent_order.recent_order).dt.days
customer_order_details = customer_order_details.merge(recent_order[['customer_id','days_recency']], on='customer_id', how='left')
customer_order_details
```

```python
df = pd.merge(customer_df, new_order, on='customer_id', how='left')
df = pd.merge(df, customer_order_details, on='customer_id', how='left')
df.info()
df
```

### Cleaning Data

terdapat sejumlah order yang belum selesai (sedang berlangsung), karena jumlahnya tidak signifikan yaitu **2.3%** sehingga dapat difilter dan menyisakan order-order yang sudah selesai dengan status **delivered** dan **canceled** 

```python
def get_minmax_scaler(df, columns):
  scaler = {}
  for col in columns:
    scaler[col] = MinMaxScaler().fit( np.array(df[col]).reshape(-1,1) )
  return scaler

def get_std_scaler(df, columns):
  scaler = {}
  for col in columns:
    scaler[col] = StandardScaler().fit( np.array(df[col]).reshape(-1,1) )
  return scaler

def apply_transform(scaler, df, prefix=None):
  temp_df = df.copy()
  for col in scaler.keys():
    temp_df[col+prefix] = scaler[col].transform( np.array(df[col]).reshape(-1,1) )
  return temp_df

def apply_log_transform(df, columns):
  temp_df = df.copy()
  for col in columns:
    temp_df[col+'_log'] = abs( np.log(temp_df[col]+1) )
  return temp_df

def zscore_remove_outliers(df, columns):
    temp_df = df.copy()
    inliers = np.array([True] * len(df))
    for col in columns:
        zscore = abs(stats.zscore(df[col]))
        inliers = (zscore < 3) & inliers
    temp_df = df[inliers]
    return temp_df


def show_boxplot(df,features):
  plt.figure(figsize=(30, 5))
  for i in range(0, len(features) ):
      plt.subplot(1, len(features) , i+1)
      sns.boxplot(y=df[features[i]], color='green', orient='v')
      plt.tight_layout()

def show_kde(df, all_columns, figsize=(30,15), yticks=[]):
  plt.figure(figsize=figsize)
  for col_name, index in zip(all_columns, range(len(all_columns))):
      plt.subplot( math.ceil(len(all_columns)/6)  , 6, index+1)
      sns.kdeplot(x=df[col_name], color='green', linewidth=3, fill=True)
      if len(yticks)>0:
        plt.yticks(yticks)
      plt.xlabel(col_name)
  plt.show()
```

```python
new_df = df.copy()
dropped_columns = ['order_purchase_timestamp','order_approved_at','order_delivered_carrier_date','order_delivered_customer_date',
                   'order_estimated_delivery_date']
new_df.drop(columns=dropped_columns,inplace=True)
new_df.info()
print('\nCount NaN Value on Data\n',new_df.isna().sum(),'\n')
new_df[(new_df.payment_sequential.isna())]
```

```python
new_df.dropna(inplace=True)
print('Count NaN value\n',new_df.isna().sum(),'\n')
new_df
```

```python
importance_features = ['days_recency','total_spending']
temp_df = new_df.copy()
temp_df = apply_log_transform(temp_df, importance_features)
show_kde(temp_df, importance_features+['days_recency_log','total_spending_log'], figsize=(30,5))
```

```python
temp_df = zscore_remove_outliers(temp_df, ['days_recency_log','total_spending_log'])
show_kde(temp_df, importance_features+['days_recency_log','total_spending_log'], figsize=(30,5))
```

```python
minmax_scaler = get_minmax_scaler(temp_df, ['days_recency_log','total_spending_log'])
temp_df = apply_transform(minmax_scaler, temp_df, prefix='_minmax')
std_scaler = get_std_scaler(temp_df, ['days_recency_log','total_spending_log'])
temp_df = apply_transform(std_scaler, temp_df, prefix='_std')

temp_df['monetary_group'] = np.where( temp_df.total_spending < temp_df.total_spending.quantile(.5), 0, np.where( temp_df.total_spending < temp_df.total_spending.quantile(.75), 1, 2))
temp_df['recency_group'] = np.ceil(temp_df.days_recency/90)-1
temp_df
```

```python
temp_df
```

```python
potential_features = ['monetary_group','recency_group']

inertia = []
for i in range(2, 11):
    k_means = KMeans(n_clusters=i, random_state=0).fit(temp_df[potential_features])
    cluster_labels = k_means.predict(temp_df[potential_features])
    inertia.append(k_means.inertia_)
    print('[ n_cluster={} ] Silhouette Score : {}'.format( i, silhouette_score( temp_df[potential_features] , cluster_labels) ))

plt.figure(figsize=(16, 8))
sns.lineplot(x=range(2, 11), y=inertia, color='#000087', linewidth = 4)
sns.scatterplot(x=range(2, 11), y=inertia, s=300, color='#800000',  linestyle='--')
plt.show()
```
